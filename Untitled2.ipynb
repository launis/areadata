```python

import pandas as pd
import numpy as np
```

```python
url="https://www.stat.fi/fi/luokitukset/corrmaps/export/kunta_1_20200101%23tyossakayntial_1_20200101/"
```

```python
f = lambda x : (x.replace("'", ""))
tkalue=pd.read_csv(url, sep=";", encoding="ISO-8859-1", skiprows=4, 
            low_memory=False, names=['Kuntanumero', 'Kunta', 'Tkaluenumero', 'Tkalue'],
            converters = {'Kuntanumero':f, 'Tkaluenumero':f})
tkalue = tkalue.apply(pd.to_numeric, errors='ignore')
```

```python
tkal = pd.DataFrame(tkalue['Tkalue'].unique(),columns=['Tkalue'])
```

```python
pd.DataFrame(tkalue['Tkalue'].unique(),columns=['Tkalue'])
```
